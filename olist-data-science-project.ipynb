**Checando estacionariedade**

Portanto para garantir a estacionariedade devemos atender a 3 pontos:

   A **media** dos valores da sua serie deve ser **constante no tempo**
   
   O **desvio padrão (variância)** desta serie também deve ser **constante no tempo**
   
   Por ultimo, **não haver componentes de sazonalidade**.
    

```python
from statsmodels.tsa.stattools import adfuller
```

```python
resultados=adfuller(vendas_mensais)

print('Teste ADF:',resultados[0])
print('p-value:',resultados[1])
```

Se **p-value>0.05** nao rejeita a hipotese nula (é uma série **não estacionária**).

```python
#Cria a funcao que faz diferenciacao
def difference(dataset,interval=1):
    diff=list()
    for i in range(interval,len(dataset)):
        value=dataset[i]-dataset[i-interval]
        diff.append(value)
  
    return diff
```

```python
#Cria a funcao que reverte o valor diferenciado para o original
def inverse_difference(history,previsao,interval=1):
    return previsao + history[-interval]
```

```python
def visualizar_residuos(modelo):
    #Armazenar os residuals
    residuals=pd.DataFrame(modelo.resid)

    #A media deve estar proximo a zero para ser um bom modelo
    print('A media deve estar proxima de zero:',residuals.describe())

    #O residual deve estar numa distribuicao normal
    residuals.plot(kind='kde')
    plt.show()
```

```python
#Transformar a serie para ter uma distribuicao normal atraves de diferenciacao, de modo a torna-la estacionaria
vendas_mensais_transformada=pd.DataFrame()
vendas_mensais_transformada['price']=difference(vendas_mensais)
vendas_mensais_transformada['order_purchase_date']=pd.DataFrame(vendas_mensais.index)
vendas_mensais_transformada.set_index('order_purchase_date',inplace=True)

vendas_mensais_transformada.head()
```

**Especificando um baseline**

* O valor previsao e o valor anterior, no nosso caso, o valor dia mes anterior

```python
#atribui os dados a uma variavel X
#X=vendas_mensais.values
X=vendas_diarias.values

#define o tamanho do conjunto de treino (60% para treino e 40% para teste)
train_size=int(len(X)*0.6)

#cria conjuntos de treino e teste
train=X[0:train_size]
test=X[train_size:]

#obtem todos os dados de treino
history=[x for x in train]

#cria lista de previsoes
predicoes=list()

#cria o modelo naive
for i in range(len(test)):
  #valor predito recebe o ultimo valor do conjunto de treino
  valor_predito=history[-1]

  #recebe o valor anterior
  predicoes.append(valor_predito)

  #valor real recebe o valor atual de teste
  valor_real=test[i]
   
  #adiciona o valor real a lista de historico
  history.append(valor_real)

  #imprime o valor predito e valor real
  #print(' Valor predito: %.2f, Valor real: %.2f'%(valor_predito,valor_real))
```

**Prevendo dados com ARIMA com Walk Forward**

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
#a variavel X recebe dados da serie
#X=vendas_mensais.values
X=vendas_diarias.values
X=X.astype('float32')
```

```python
#obtem todos os dados de treino
history=[x for x in train]

#cria lista de previsoes
predicoes=list()

#Inicia o Walk-Foward
for t in range(len(test)):
    meses_no_ano=1
    
    diff=difference(history,meses_no_ano)
  
    #cria o modelo arima com dados de history (Auto Regressivo, Diferenciacao, Media Movel)
    model=ARIMA(diff,order=(1,0,0))
  
    #treina o modelo ARIMA
    model_fit=model.fit(trend='nc',disp=0)

    #Prever
    valor_predito=model_fit.forecast()[0]

    #Reverter o valor para escala original
    valor_predito=inverse_difference(history, valor_predito,meses_no_ano)

    #Adicionafr ana lista de predicoes
    predicoes.append(valor_predito)

    #valor real recebe o valor atual
    valor_real=test[t]

    #adicionar o valor real ao history
    history.append(valor_real)

    #imprime o valor predito e valor real
    #print('Valor predito=%.3f, Valor esperado=%.3f'%(valor_predito,valor_real))
```

```python
#Avaliando os resultados
rmse=sqrt(mean_squared_error(predicoes,test))
print('ARIMA - Teste RMSE: %.3f'%rmse)
```

Esta acima do taxa de acerto do modelo base

```python
#Visualizar os erros
visualizar_residuos(model_fit)
```

Resumo:

**Distribuicao assimetrica postiva** e media distante de zero (nao e um bom modelo).

Um bom modelo deve ter uma **distribuicao normal** nos seus residuos.
